```python
# 使用条件表达式构造布尔列表
条件1 = df3.open < 35
条件2 = df3.close < 35
筛选条件 = 条件1 & 条件2
df3.loc[筛选条件]
```

```python
# 使用函数作为索引方式
def 筛选函数(df):
    条件1 = df.open < 35
    条件2 = df.close < 35
    筛选条件 = 条件1 & 条件2
    return 筛选条件
```

```python
df3.loc[筛选函数]
```

```python
筛选函数(df3)
```

```python
# 使用Query方法查询数据
df3.query('open < 35 and close < 35')
```
